## Data Preprocessing

### Spliting data into test and train

```python
# Creating Variable to store feature 
X = cancer_df.drop(columns='target',axis= "columns")
```

```python
X.head()
```

```python
# creating the Variable to store the target Variable
Y = cancer_df['target']
```

```python
Y.head()
```

```python
from sklearn.model_selection  import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = .2,random_state = 5)

X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

### Feature Sacling

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train_sc = scale.fit_transform(X_train)
X_test_sc = scale.transform(X_test)
```

## Now which M/C Learning Algo to use.

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

### SVM:
### support vector machine (SVM) is a supervised machine learning model that uses classification algorithms for two-group classification problems. After giving an SVM model sets of labeled training data for each category, they're able to categorize new text

```python
# Support vector classifier
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train, Y_train)
Y_pred_scv = svc_classifier.predict(X_test)
accuracy_score(Y_test, Y_pred_scv)
```

```python
# With Scaled Data
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train_sc,Y_train)
Y_prediction_svc_sc = svc_classifier.predict(X_test_sc)
print(accuracy_score(Y_test,Y_prediction_svc_sc))
```

## Logistic regression
####Logistic Regression, also known as Logit Regression or Logit Model, is a mathematical model used in statistics to estimate (guess) the probability of an event occurring having been given some previous data. Logistic Regression works with binary data, where either the event happens (1) or the event does not happen (0).

```python
# with Unscaled Data
from sklearn.linear_model import LogisticRegression
lr_classifire = LogisticRegression(random_state = 51, penalty = 'l2')
lr_classifire.fit(X_train,Y_train)
Y_prediction_lr = lr_classifire.predict(X_test)
print(accuracy_score(Y_test,Y_prediction_lr))
```

```python
from sklearn.linear_model import LogisticRegression
lr_classifire = LogisticRegression(random_state = 51, penalty = 'l2')
lr_classifire.fit(X_train_sc,Y_train)
Y_prediction_lr_sc = lr_classifire.predict(X_test_sc)
print(accuracy_score(Y_test,Y_prediction_lr_sc))
```

## K Neighbors Classifier
K Nearest neighbors is a simple algorithm that stores all available cases and classifies new cases based on a similarity measure (e.g., distance functions). KNN has been used in statistical estimation and pattern recognition already in the beginning of 1970's as a non-parametric technique.

```python
#KNN with the Scaled Data
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski',p = 2)
knn_classifier.fit(X_train_sc,Y_train)
Y_prediction_sc = knn_classifier.predict(X_test_sc)
print(accuracy_score(Y_test,Y_prediction_sc))
```

```python
#KNN with the Scaled Data
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski',p = 2)
knn_classifier.fit(X_train,Y_train)
Y_prediction = knn_classifier.predict(X_test)
print(accuracy_score(Y_test,Y_prediction))
```

# Naive Bayes Classifire
# Probabilistic model
##stractly, naÃ¯ve Bayes is a conditional probability model: given a problem instance to be classified, represented by a vector representing some n features (independent variables), it assigns to this instance probabilities. for each of K possible outcomes or classes .

```python
# With Scaled Data
from sklearn.naive_bayes import GaussianNB
NB_classifier = GaussianNB()
NB_classifier.fit(X_train_sc,Y_train)
Y_prediction = NB_classifier.predict(X_test_sc)
print(accuracy_score(Y_prediction,Y_test))
```

```python
# With UnScaled data
from sklearn.naive_bayes import GaussianNB
NB_classifier = GaussianNB()
NB_classifier.fit(X_train,Y_train)
Y_prediction = NB_classifier.predict(X_test)
print(accuracy_score(Y_test,Y_prediction))
```

## Decision Tree Classifier
Decision tree builds classification or regression models in the form of a tree structure. It breaks down a data set into smaller and smaller subsets while at the same time an associated decision tree is incrementally developed. The final result is a tree with decision nodes and leaf nodes.

```python
# Decision Tree with Scaled Data...

from sklearn.tree import DecisionTreeClassifier
DT_classifier = DecisionTreeClassifier(criterion='entropy')
DT_classifier.fit(X_train_sc,Y_train)
Y_prediction = DT_classifier.predict(X_test_sc)
print(accuracy_score(Y_prediction,Y_test))
```

```python
# Decision Tree with UnScaled Data...


DT_classifier2 = DecisionTreeClassifier(criterion='entropy')
DT_classifier2.fit(X_train_sc,Y_train)
Y_prediction = DT_classifier2.predict(X_test_sc)
print(accuracy_score(Y_prediction,Y_test))
```

# Random Forest Classifier
#Random forest is a supervised learning algorithm which is used for both classification as well as regression. ... Similarly, random forest algorithm creates decision trees on data samples and then gets the prediction from each of them and finally selects the best solution by means of voting

```python
# Random Forest Classifier on Scaled Data
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'entropy')
rf_classifier.fit(X_train_sc, Y_train)
y_pred_rf = rf_classifier.predict(X_test_sc)
accuracy_score(Y_test, y_pred_rf)
```

```python
# Random Forest Classifier on Unscaled Data
from sklearn.ensemble import RandomForestClassifier
rf_classifier2 = RandomForestClassifier(n_estimators = 20, criterion = 'entropy')
rf_classifier2.fit(X_train, Y_train)
y_pred = rf_classifier2.predict(X_test)
accuracy_score(Y_test, y_pred)
```

# XGBoost Classifier

#### XGBoost is an optimized distributed gradient boosting library designed to be highly efficient, flexible and portable. It implements machine learning algorithms under the Gradient Boosting framework. XGBoost provides a parallel tree boosting (also known as GBDT, GBM) that solve many data science problems in a fast and accurate way.

```python
# XGBoost with Scaled Data..

from xgboost import XGBClassifier
xgb_classifier = XGBClassifier()
xgb_classifier.fit(X_train_sc,Y_train)
Y_prediction_xgb_sc = xgb_classifier.predict(X_test_sc)
print(accuracy_score(Y_test,Y_prediction_xgb_sc))
```

```python
# XGBoost with UnScaled Data..

from xgboost import XGBClassifier
xgb_classifier1 = XGBClassifier()
xgb_classifier1.fit(X_train,Y_train)
Y_prediction_xgb = xgb_classifier1.predict(X_test)
print(accuracy_score(Y_test,Y_prediction_xgb))
```

## XGBoost Parameter Tunning (Randomised Search)

```python
# XGBoost classifier most required parameters
params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ] 
}
```

```python
# Randomized Search
from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(xgb_classifier, param_distributions=params, scoring= 'roc_auc', n_jobs= -1, verbose= 3)
random_search.fit(X_train, Y_train)
```

# Confusion Matrix

```python
cm = confusion_matrix(Y_test, Y_prediction_xgb_sc)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sbs.heatmap(cm, annot = True)
plt.show()
```

## Print Classification Report

```python
print(classification_report(Y_test,Y_prediction_xgb_sc))
```
